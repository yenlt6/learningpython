```python
class Square:
    def __init__(self, length):
        self._length = length
    def __str__(self):
        return (f"Square: {self._length}")

    def chuvi(self):
        return self._length*4

    def dientich(self):
        return self._length*self._length 

    
class Rectangular:
    
    def __init__(self, width, height):
        self._width = width
        self._height = height
   
    def __str__(self):
        return (f"Rectangular: {self._width}, {self._height}")

    def chuvi(self):
        return (self._width+ self._height)*2

    def dientich(self):
        return self._width*self._height


list_obj = []
for i in range (5):
    list_obj.append(Square((i+1)*5))
    list_obj.append(Rectangular((i+1)*5, (i+1)*6))

for obj in  list_obj:    
    print (obj, "Chu vi:",obj.chuvi(), "Diện tích", obj.dientich())   
    
```
